```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from typing import Set, Any
from tkinter import *
from PIL import Image, ImageTk
fooditem = " "
piefile = " "
barfile = " "
```

```python
def remove_others(df: pd.DataFrame, columns: Set[Any]):
    cols_total: Set[Any] = set(df.columns)
    diff: Set[Any] = cols_total - columns
    df.drop(diff, axis=1, inplace=True)
data = pd.read_csv("C:/Users/geetika singh/Desktop/epi_r.csv",index_col=None)
remove_others(data,{"title",'rating','calories','protein','fat','sodium','bread','blue cheese','boil','breakfast'
                        ,'brown rice','brunch','cheese','chicken','coffee','dairy','dairy free','dinner',
                        'egg','fat free','fish','healthy','low carb','low fat','low cal','low cholesterol','low sodium',
                         'low sugar','low/no sugar','lunch','milk/cream','oatmeal','pasta','salad','salmon',
                        'sandwich','slow cook','shrimp','soy free','vegetarian','yogurt','zucchini'})
```

```python
print(data.shape)
# drop rows with missing values
data.dropna(inplace=True)
print(data.shape)
```

```python
items = list(data.title)
print(items)
```

```python
def showPieChart(fooditem):
    global piefile
    labels = 'Calories', 'Protein', 'Fat', 'Sodium'
    value_row = np.array(data.loc[data["title"]==fooditem])
    values = value_row[0,2:6]
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
    explode = (0.1, 0, 0, 0)  # explode 1st slice
    fig = plt.figure(figsize=(6, 3))
    plt.pie(values, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=100)
    plt.axis('equal')
    plt.title(fooditem)
    piefile = fooditem+'pie.png'
    fig.savefig(piefile)
    plt.close()
```

```python
def showBarGraph(fooditem):
    global barfile
    plt.rcdefaults()
    value_row = np.array(data.loc[data["title"]==fooditem])
    values = value_row[0,2:6]
    objects = ('Cal', 'Prot', 'Fat', 'Sod')
    y_pos = np.arange(len(objects))
    performance = values
    fig = plt.figure(figsize=(6, 4))
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.xlabel('nutrients')
    plt.ylabel('value')
    plt.title(fooditem)
    barfile = fooditem+'bar.png'
    fig.savefig(barfile)
    plt.close()
```

```python
def get_list(event):
    """
    function to read the listbox selection
    and put the result in a label
    """
    # get line index tuple
    sel = listbox1.curselection()
    # get the text
    seltext = listbox1.get(int(sel[0]))
    label1.config(text=seltext)
    global fooditem
    fooditem = seltext
    showPieChart(fooditem)
    load1 = Image.open(piefile)
    render1 = ImageTk.PhotoImage(load1)
    img1 = Label(w, image=render1)
    img1.image = render1
    img1.grid(row=0,column=0)
    showBarGraph(fooditem)
    load2 = Image.open(barfile)
    render2 = ImageTk.PhotoImage(load2)
    img2 = Label(w, image=render2)
    img2.image = render2
    img2.grid(row=1,column=0)
    
root = Tk()
root.state('zoomed')
listbox1 = Listbox(root, width=80, height=20)
listbox1.grid(row=0, column=0)
# create a vertical scrollbar to the right of the listbox
yscroll = Scrollbar(command=listbox1.yview, orient=VERTICAL)
yscroll.grid(row=0, column=1, sticky=N+S)
listbox1.configure(yscrollcommand=yscroll.set)
#creating a canvas to display graphs
w = Canvas(root,width=500,height=250)
w.grid(row=0,column=2)
# label to display selection
label1 = Label(root, text='Click on an item in the list',font=('Pursia',10,'italic'))
label1.grid(row=1, column=0)
# load the listbox with data
for item in items:
    listbox1.insert(END, item)
# left mouse click on a list item to display selection
listbox1.bind('<ButtonRelease-1>', get_list)
root.mainloop()
```
